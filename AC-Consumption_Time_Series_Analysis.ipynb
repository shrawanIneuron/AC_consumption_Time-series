```python
# Doing Hypothesis test to check whether data is stationary or not

#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(Total_Consumption):
    result=adfuller(Total_Consumption)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
    
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfuller_test(df_minutes['Total consumption'])
```

```python
#Determing rolling statistics
rolLmean = df_minutes.rolling(12).mean()
rolLstd = df_minutes.rolling(12).std()

plt.figure(figsize=(16,7))
fig = plt.figure(1)

#Plot rolling statistics:
orig = plt.plot(df_minutes['Total consumption'], color='blue',label='Original')
mean = plt.plot(rolLmean, color='red', label='Rolling Mean')
std = plt.plot(rolLstd, color='black', label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

```python
#Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_minutes['Total consumption'],freq=1,model = 'additive')

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(16,7))
fig = plt.figure(1)

plt.subplot(411)
plt.plot(df_minutes['Total consumption'], label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual, label='Residuals')
plt.legend(loc='best')
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(df_minutes['Total consumption'], nlags=32)
lag_pacf = pacf(df_minutes['Total consumption'], nlags=32)
```

```python
import statsmodels.api as sm
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(df_minutes['Total consumption'].dropna(),lags=32,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(df_minutes['Total consumption'].dropna(),lags=32,ax=ax2)
```

```python
# Building ARIMA model for prediction
from statsmodels.tsa.arima_model import ARIMA
```

```python
type(df_minutes['Total consumption'])
```

```python
# p=2,q=3,d=0 from the above acf and pacf graph
plt.figure(figsize=(16,8))
model = ARIMA(df_minutes['Total consumption'], order=(2,0,3))  
results_ARIMA = model.fit()  
plt.plot(df_minutes['Total consumption'])
plt.plot(results_ARIMA.fittedvalues, color='red')
```

```python
results_ARIMA.summary()
```

```python
df_minutes['forecast']=results_ARIMA.predict(start=5,end=34000,dynamic=True)
df_minutes[['Total consumption','forecast']].plot(figsize=(12,8))
```
